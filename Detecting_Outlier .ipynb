# Z-score

```python
from scipy import stats
z = np.abs(stats.zscore(df))
z
```

```python
threshold = 3 # identifying outliers
np.where(z>threshold)
```

```python
z[55][2]
```

# Cleaning Data by Removing Outliers

```python
df_no_outlier = df[(z<3).all(axis=1)]
df_no_outlier
```

```python
df_no_outlier.drop('Unnamed: 0',axis=1,inplace=True)
df_no_outlier
```

# Finding Outliers Using IQR

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 - q1
IQR
```

```python
boston_IQR_Outlier = df[~((df<(q1 - 1.5 * IQR))|(df>(q3 + 1.5 * IQR))).any(axis=1)]
boston_IQR_Outlier
```
