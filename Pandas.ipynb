# Groupby for Categorical Columns

```python
dataset = {'Company':["Google","Google","Microsoft","Instagram","Microsoft","Twitter","Facebook"],
          'Names':["Gagan","Varun","Vishal","Ram","Shyam","Gowri","Manoj"],
           'Sales':[200,450,350,600,120,540,770]}
dataset = pd.DataFrame(dataset)
dataset.head()
```

```python
bycomp = dataset.groupby('Company')# applied only on categorical variables
bycomp
```

```python
bycomp.mean()
```

```python
bycomp.max()
```

```python
bycomp.describe()
```
