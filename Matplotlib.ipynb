# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
print(x)
print(y)
```

```python
plt.plot(x,y,'r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("X vs Y")
plt.show()
```

# Multiple plots on same canvas

```python
plt.subplot(2,2,1)
#plt.plot(x,y,'r',marker='*')
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(y,x,'g-')
plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
plt.subplot(2,2,4)
plt.plot(y,x,'y+-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'g')
axes1.set_xlabel("x-main")
axes1.set_ylabel("y-main")
axes1.set_title("Main Graph")

axes2.plot(y,x,'b*')
axes2.set_xlabel("x-small")
axes2.set_ylabel("y-small")
axes2.set_title("Sub Graph")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label = 'x**2',color='m')
ax.plot(x,x**3,label='x**3',color='b')
ax.legend()

#ax.legend(loc=1) upper right corner
#ax.lengend(loc=2) upper left corner
#ax.legend(loc=3) lower left corner
#ax.legend(loc=4) lower right corner
```
