```python
import pandas as pd
import numpy as np
```

# Creating Time Series

```python
time_series = pd.date_range('1/1/2020',periods=20) 
df = pd.DataFrame(np.random.randn(20,5),index=time_series,columns=["Column 1","Column 2","Column 3","Column 4","Column 5"])# 20 rows 5 columns
df.head()
```

```python
df['Column 1']
```

```python
df[['Column 1','Column 2','Column 3']]
```

# Time series specifying index (Indexing)

```python
col1 = df['Column 1']
col1[time_series[3]]
```

```python
# dataframe.loc attribute access a group of rows or columns by labels in the given dataframe
```

```python
x = pd.DataFrame({'x':[1,2,3,4],'y':[5,6,7,8]})
# selecting by position
print(x)
# specifying new values to the row number 1
x.loc[1] = {'x':9,'y':99}
print('\n')
print(x)
```

# Slicing

```python
df.iloc[:5,:2] # iloc attribute access a group of rows or columns by the index location
```

```python
df[::3] # all columns with every thrid row and not all the rows
```

# Filtering - filters the given sequence and returns it

```python
df[(df['Column 3'] < 0)]
```

```python
df[(df['Column 1']< 0) & (df['Column 2'] > 0)][['Column 4','Column 5']]
```
